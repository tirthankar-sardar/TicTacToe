```python
from random import randint

def display_board():
  print()
  for x in range(3):
    line = " "
    for y in range(3):
      line += board[x][y] + " | "*(y<2)
    print(line + "\n---+---+---"*(x<2))
  print()

def two_on_a_line(letter):
  lines = []
  for n in range(3):
    lines.append(board[n][0] + board[n][1] + board[n][2])
    lines.append(board[0][n] + board[1][n] + board[2][n])

  lines.append(board[0][0] + board[1][1] + board[2][2])
  lines.append(board[0][2] + board[1][1] + board[2][0])

  for line in lines:
    if line.count(letter) == 2:
      other = line.replace(letter,"")
      if other in valid:
        return other
  return ""

def has_won(letter):
  global finished
  for n in range(3):
    if board[n][0] == board[n][1] == board[n][2] == letter or board[0][n] == board[1][n] == board[2][n] == letter:

      finished = True

    if board[0][0] == board[1][1] == board[2][2] == letter or board[0][2] == board[1][1] == board[2][0] == letter:
      finished = True
  return finished

def user_go():
  square = ""
  while square not in valid:
    square = input("Select a numbered square from the board and press Enter: ")

  valid.pop(valid.index(square))
  square = int(square)-1
  board[square // 3][square % 3] = player[1-PC]

  display_board()

  if has_won(player[1-PC]):
    print("You win - well done!\n")

def computer_go(move):
  corner = (0,2)
  stuck = False

  if move == 1:

    x = corner[randint(0,1)]
    y = corner[randint(0,1)]

    while board[x][y] not in valid:
      x = corner[randint(0,1)]
      y = corner[randint(0,1)]

    square = board[x][y]

    valid.pop(valid.index(square))

    board[x][y] = player[PC]

  if move == 2 and PC == 0:
    opposite = False

    for x in corner:
      for y in corner:

        oppx = abs(x-2)
        oppy = abs(y-2)
        if board[x][y] == player[PC] and board[oppx][oppy] in valid:
          square = board[oppx][oppy]
          valid.pop(valid.index(square))
          board[oppx][oppy] = player[PC]
          opposite = True

    if not opposite:
      while board[x][y] not in valid:
        x = corner[randint(0,1)]
        y = corner[randint(0,1)]
      square = board[x][y]
      valid.pop(valid.index(square))
      board[x][y] = player[PC]

  if move in (3,4) or (move == 2 and PC == 1):

    square = two_on_a_line(player[PC])
    if square == "":

      square = two_on_a_line(player[1-PC])

    if square == "":
      x = y = 0

      if len((board[0][0]+board[0][2]+board[2][0]+board[2][2]).replace("X","").replace("O","")) == 0:
        stuck = True
      while board[x][y] not in valid and not stuck:
        x = corner[randint(0,1)]
        y = corner[randint(0,1)]
      square = board[x][y]

    if not stuck:
      valid.pop(valid.index(square))
      square = int(square)-1
      board[square // 3][square % 3] = player[PC]

  if move == 5 or stuck:

    done = False
    for x in range(3):
      for y in range(3):
        square = board[x][y]
        if square in valid and not done:
          valid.pop(valid.index(square))
          board[x][y] = player[PC]
          done = True

  display_board()

  if has_won(player[PC]):
    print("The computer wins - bad luck!\n")


board = [["1","2","3"],["4","5","6"],["7","8","9"]]
valid = ["1","2","3","4","5","6","7","8","9"]

player = ["X","O"]
PC = -1
while PC not in ("0","1"):
  PC = input("\nWho goes first?  Enter 0 for the computer and 1 for you: ")
PC = int(PC)
go = 1
finished = False
print("\nYou are",player[1-PC],"and the computer is",player[PC]+"."+"\n"*(PC==0))

if PC == 1:
  display_board()

while go < 10 and finished == False:

  if go % 2 == PC:
    user_go()
  else:
    print("Computer's go...")
    computer_go(int((go+(1-PC))/2))
  go += 1
if not finished:
  print("It's a draw!")
```
